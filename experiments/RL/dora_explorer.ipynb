```python
import pickle
import matplotlib.pyplot as plt
import os
import numpy as np
```

```python
wds = [1,1e-2,1e-4,1e-6,0.0]

for basic_wd in wds:
    for equiv_wd in wds:
        fname = "MixedEMLP_basic" + str(basic_wd) + "_equiv" + str(equiv_wd)
        fname = "./saved-outputs/" + fname + "epoch_rewards.pkl"
        if os.path.exists(fname):
            filehandler = open(fname,"rb")
            rewards = pickle.load(filehandler)
            filehandler.close()
            print("basic_wd = ", basic_wd)
            print("equiv_wd = ", equiv_wd)
            print(rewards[-10:])
            print(len(rewards))
            print("\n")
            
```
