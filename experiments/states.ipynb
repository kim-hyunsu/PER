```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from pandas import DataFrame
import csv
```

```python
# state selection figure
# fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(12,4) ,constrained_layout=True)
file = pd.read_csv('./state.csv')
data = pd.DataFrame(file)
state = data['State'].tolist()
plt.xticks(fontsize =13)
plt.yticks(fontsize =13)
plt.rcParams['figure.figsize'] = 10, 6
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42
plt.plot(list(range(0,10000)), state, label='state')
plt.legend(prop={'size':20})
# plt.show()
plt.savefig('./states.pdf', dpi=10000)
```

```python
file = pd.read_csv('./states.csv')
data = pd.DataFrame(file)
data = data.replace(0, np.nan)
data = data.dropna(axis=0)
# print(data)
state = data['State'].tolist() # state-1
state0 = data['State0'].tolist()
state1 = data['State1'].tolist()
state2 = data['State2'].tolist()
state3 = data['State3'].tolist()
state4 = data['State4'].tolist()
state5 = data['State5'].tolist()
state6 = data['State6'].tolist()
state7 = data['State7'].tolist()
state8 = data['State8'].tolist()
print(len(state))
```

```python
# Big one
# Make data
data = pd.DataFrame({'state_0':state0, 'state_1':state1, 'state_2':state2,
'state_4':state4, 'state_5':state5}, index=range(0,199))
 
# We need to transform the data from raw data to percentage (fraction)
data_perc = data.divide(data.sum(axis=1), axis=0)
 
# Make the plot
plt.stackplot(range(0,9950,50),  data_perc["state_0"], data_perc["state_1"], data_perc["state_2"], 
data_perc["state_4"], data_perc["state_5"], labels=['SE3','R3','T3','Ryz2','Rxz2'], )
plt.legend(loc='upper left', fontsize=16)
# plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.margins(0,0)
plt.xticks(fontsize =20)
plt.yticks(fontsize =20)
plt.rcParams['figure.figsize'] = 10, 6
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42
plt.savefig('./rxy2_big.pdf', dpi=10000)
plt.show()
```

```python
# small one
# Make data
data = pd.DataFrame({'state_-1':state, 'state_3':state3, 'state_6':state6,
'state_7':state7, 'state_8':state8}, index=range(0,199))
 
# We need to transform the data from raw data to percentage (fraction)
data_perc = data.divide(data.sum(axis=1), axis=0)
 
# Make the plot
plt.stackplot(range(0,9950,50),  data_perc["state_-1"], data_perc["state_3"], data_perc["state_6"], 
data_perc["state_7"], data_perc["state_8"], labels=['No PER','Rxy2','Txy2','Tyz2','Txz2'], 
colors=['gold', 'lightcyan', 'silver', 'mistyrose', 'burlywood'],)
plt.legend(loc='upper left', fontsize=16)
plt.margins(0,0)
plt.xticks(fontsize =20)
plt.yticks(fontsize =20)
plt.rcParams['figure.figsize'] = 10, 6
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42
plt.savefig('./rx2_small.pdf', dpi=10000)
plt.show()
```
